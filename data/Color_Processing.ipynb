```python
im_arr_mscoco = []

for counter, image in enumerate(os.listdir()):
    im_arr_mscoco.append(np.asarray(Image.open(image)))
    print('processing image{}'.format(counter))
    if counter == 10000:
        break
```

```python
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Resize((224, 224))])
```

```python
im_transform = []
for img in im_arr_mscoco:
     im_transform.append(transform(img).unsqueeze(3))
```

```python
im_transform[0].shape != (3,224,224,1)
```

```python
im_transform[0].shape
```

```python
idx_list = []
for idx, im in enumerate(im_transform):
    if im.shape != (3,224,224,1):
        del im_transform[idx]
        idx_list.append(idx)
```
